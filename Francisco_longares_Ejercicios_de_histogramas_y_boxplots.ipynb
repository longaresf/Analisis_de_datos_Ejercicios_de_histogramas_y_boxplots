##Ejercicios de histogramas y boxplots

```python
# Librerias a importar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importar DataFrame
path = '/content/drive/MyDrive/Colab Notebooks/dataset_1/2015.csv'
df_csv = pd.read_csv(path)
# 1. Cargar y Explorar los Datos
# Imprimir los 5 primeros datos
print(df_csv.head())
```

```python
# Imprimir los nombres de las columnas para verificar
print(df_csv.columns)
```

##2. Crear Histogramas:

```python
variables_to_plot = ('Happiness Rank', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom')
```

```python
for var in variables_to_plot:
  plt.figure(figsize=(10, 6))
  sns.histplot(data=df_csv, x=var, kde=True, color='skyblue', edgecolor='black', alpha=0.7)
  plt.title(f'Histograma de {var}')
  plt.xlabel(var)
  plt.ylabel('Frecuencia')
  plt.show()
```

##3. Crear Boxplots:

```python
# Crear boxplots
for var in variables_to_plot:
  plt.figure(figsize=(10, 6))
  sns.boxplot(data=df_csv, x=var, color='lightgreen', showfliers=True)
  plt.title(f'Boxplot de {var}')
  plt.xlabel(var)
  plt.show()
```

##4. Anlisis Combinado:

```python
# Anlisis Combinado
for var in variables_to_plot:
  plt.figure(figsize=(12, 8))

  # Histograma
  plt.subplot(2, 1, 1)
  sns.histplot(data=df_csv, x=var, kde=True, color='skyblue', edgecolor='black', alpha=0.7)
  plt.title(f'Histograma y Boxplot de {var}')
  plt.xlabel(var)
  plt.ylabel('Frecuencia')
```
